## Data Cleaning ##

```python
def clean_price(price):
    if isinstance(price,str):
        if '£' in price:
            price = price.replace('£','').replace('each est.','').replace('-','').strip()
        elif 'p' in price:
            price = '0.' + price.replace('p','').replace('each est.','').replace('-','').strip()

    return float(price)
```

```python
term_df['price'] = term_df['item-price'].apply(clean_price)
```

## specific Product ##

```python
salt = term_df[
    term_df['name'].str.contains('salted butter', case=False, na=False) & 
     (term_df['category'].str.contains('everyday', case=False, na=False) | 
    ~term_df['category'].str.contains('everyday', case=False, na=False))
]
```

```python
salt_remove = ["Naturli Organic Vegan Unsalted Butter Alternative",
                "Essential Unsalted Butter",
                "Duchy Organic English Unsalted Butter",
                "Flora Plant Unsalted Butter Alternative",
                "Yeo Valley Organic Unsalted Butter",
                "President Unsalted Butter",
                "Country Life Unsalted Butter",
                "Lurpak Unsalted Butter",
                "Lescure French Unsalted Butter",
                "Waitrose Salted Butterscotch Dairy Ice Cream"]

salted_butter = salt[~salt['name'].str.contains('|'.join(salt_remove), case=False, na=False)]
```

```python
salt.shape
```

```python
salt_plot = (
    salted_butter.assign(new_price = lambda s: s['price'] / s['size'].str.replace('g','').astype('float') * 1000,
    price= lambda x: x['price'],
    size = lambda x: x['size'])
)


salt_plot[['name','price','size','new_price']]
```

```python
(
    ggplot(salt_plot, aes(x='name', y='new_price')) +
    geom_line(color='red', size=1.2) +
    labs(
        title='Price of Saltedbutter ',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)
)
```

```python
salt_plot = salt_plot.sort_values(by='price', ascending=False)
tooltip_setup = (
    layer_tooltips()
        .line('@name')
        .line('[@size -- @price')
        .format('@price', '£ {.2f}')
)

salt_plots= (
    # Maps the columns to the aesthetics of the plot.
    ggplot(salt_plot, aes(y='name', x='price' ,fill='name')) +

    # GEOMS

    # Add a line range that 'listens to' columns informed in `ymin` and `ymax` aesthetics
    #geom_linerange(size=1, alpha=0.75, tooltips=tooltip_setup) +

    # Add points to the plot (listen to `x` and `y` and fill aesthetics)
    geom_point(size=3, stroke=1, shape=21, tooltips=tooltip_setup) +

    # SCALES

    # Remove the legend (we can already read the categories from the y-axis)
    scale_fill_discrete(guide='none') +

    # Specify names for the axes
    scale_y_continuous(name="Categories\n(from largest to smallest median)", expand=[0.05, 0.05]) +
    scale_x_continuous(name="Price (£)", expand=[0., 0.05], format='£ {.2f}', breaks=np.arange(0, 20, 0.5)) +

    # LABELS
    # It's nice when the plot tells you the key takeaways
    labs(title='"Beer, Wine & Spirits" has the highest median price for individual items',
         subtitle="Dots represent the median price, bars represent the 25th and 75th percentiles") +
    theme(axis_text_x=element_text(size=10),
        axis_text_y=element_text(size=17),
        axis_title_x=element_text(size=20),
        axis_title_y=element_text(size=20),
        plot_title=element_text(size=19, face='bold'),
        plot_subtitle=element_text(size=18),
        legend_position='none') +
    ggsize(1200, 500)

)
salt_plots
```
